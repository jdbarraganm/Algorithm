```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from time import time
```

```python
def fibo(n):
    
    a = 0
    b = 1
    i = 2  
    if(n==0):
        return n
    else:
        while i <= n:
            b += a
            a = b - a   
            i += 1
        return b
```

```python
excTime = []
total = 0.0
for i in range(3000):
    start_time = time()
    fibo(i)
    end_time = time()
    total += (end_time - start_time)
    excTime.append(total)
```

```python
plt.plot(excTime,np.arange(3000));
plt.axis()
plt.xlabel('Time(s)')
plt.ylabel('N')
plt.title('Execution Time to calculate Fibonacci(n)');
```
