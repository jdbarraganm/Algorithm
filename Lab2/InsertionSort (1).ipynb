## Número de instrucciones

```python
import matplotlib.pyplot as plt
import numpy as np

npermutacion = 0
total = 0
casos = np.zeros(1001)
prob = np.zeros(1001)
minimo = 1000
maximo = 0

def sortSteps(a):
    v = []
    i = 0
    while (i<len(a)):
        v.append(a[i])
        
    pasos = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            pasos = pasos + 3
        pasos = pasos + 1
        v[j+1] = x
        pasos = pasos + 4
    pasos = pasos + 1
    return pasos

def swap(v,i,j):
    t = v[i]
    v[i] = v[j]
    v[j] = t

def permutar(v,i,n):
    global npermutacion
    global total
    global casos
    
    j = 0
    k = 0
    l = 0
    if i==n:
        npermutacion = npermutacion +1
        k = sortSteps(v)
        total = total + k
        casos[k] = casos[k] + 1
    else:
        for j in range(i,n):
            swap(v,i,j)
            permutar(v,i+1,n)
            swap(v,i,j)
            
n=2
v=[]
for i in range(n):
    v.append(i+1)
    
permutar(v,0,len(v))

for k in range(1001):
    prob[k] = casos[k]/float(npermutacion)
    if(casos[k]!=0) & (minimo > k):
        minimo = k
    if(casos[k]!=0) & (maximo < k):
        maximo = k
        
plt.bar(range(minimo,maximo+1),prob[minimo:maximo+1])
plt.xlabel('Lineas ejecutadas')
plt.ylabel('Probabilidad')
plt.title('Frecuencia lineas ejecutadas')
print ('Promedio') ,(float(total)/npermutacion)
```
