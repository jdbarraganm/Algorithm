```python
import matplotlib.pyplot as plt
from itertools import permutations
import random
class Counter:
    def __init__(self):
        self.steps = 0

    def reset(self):
        self.steps = 0

    def count(self):
        self.steps += 1

    def print_steps(self):
        return self.steps
        
def insertionsort(A,swap):   
    for j in range(1,len(A)):
        key = A[j] #look
        i = j-1 
        while (i > -1) and key < A[i]: 
            A[i+1]=A[i]
            swap.count()
            i=i-1
        A[i+1] = key
    return A
    
n= 5
s= random.sample(range(100), n)
s = list (s)
x= list (permutations(s, n))
f=n
fac=1
p=[]
while f > 0:
    fac= fac*f
    f= f-1
for i in range (0,fac):
    t = list (x[i])
    swap = Counter ()
    insertionsort(t,swap)
    p.append(swap.print_steps())
    
prom=0
suma = 0    
for h in range (0,fac):
    suma = suma + p[h]
    
prom = (suma)/fac
print ("el promedio es : " , prom)
print (p)
plt.hist(p,fac,(min(p),max(p)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

npermut = 0
total = 0
cases = np.zeros(1001)
prob = np.zeros(1001)
minCase = 1000
maxCase = 0


def isortSteps(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            steps = steps + 3
        steps =+ 1
        v[j+1] = x
        steps = steps + 4
    steps =+ 1
    return steps

def swap(v,i,j):
    t = v[i]
    v[i] = v[j]
    v[j] = t

def perm(v,i,n):
    global npermut
    global total
    global cases
    
    j=0
    k=0
    l=0
    if i == n:
        npermut =+ 1
        k = isortSteps(v)
        total = total + k
        cases[k] = cases[k] + 1
    else:
        for j in range (i,n):
            swap (v, i, j)
            perm (v, i+1, n)
            swap (v, i, j)

n=5
v=[]
for i in range(n):
    v.append(i+1)
    
perm(v,0,len(v))

for k in range(1001):
    prob[k] = cases[k]/float(npermut)
    if (cases[k]!=0) & (minCase > k):
        minCase = k
    if (cases[k]!=0) & (maxCase < k):
        maxCase = k

for k in range(minCase,maxCase+1):
    print ('prob(',k,')' ,prob[k])
        
plt.bar(range(minCase,maxCase+1),prob[minCase:maxCase+1])
plt.xlabel('Lineas')
plt.ylabel('Probabilidad')
plt.title('Frecuencia del numero de lineas ejecutadas');
```
